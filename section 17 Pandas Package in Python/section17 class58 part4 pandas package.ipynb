```python
# I want to create a new column for male, female as m or f.

df['Gender_Letter'] = df.Gender.map({'male': 'M', 'female': 'F'})
```

```python
df['can_vote'] = df.Age.astype(int) >= 18
```

```python
#I want to convert the colors to number as per alphabetical wise. categorical data to number.
df['color_code'] = df.color.factorize()[0]
```
