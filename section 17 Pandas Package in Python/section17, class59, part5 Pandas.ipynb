```python
df = pd.DataFrame({'orderid': [1,2,3,4,5,6],
             'Manufacturer': ['A', 'A', 'B', 'C', 'D', 'E'],
             'shipmode': ['Same Day', '1st Class', '2nd Class', 'Standard Class', 'Same Day', '2nd Class']})
```

```python
df
```

```python
from pandas.api.types import CategoricalDtype
```

```python
type = CategoricalDtype(['Standard Class', '2nd Class', '1st Class', 'Same Day'], ordered = True)
```

```python
type
```

```python
#Here i want to put this ranking which is 'standard class' is poor, '2nd class' is average, '1st class' is good and 
# 'same day ' is best the ranking.

# In type we have passed the ranking and this ranking we will put in ship mode.

df['shipmode'] = df.shipmode.astype(type)
```

```python
df.info()
```

```python
df.shipmode #It clearly states the ranking as standard class is worst and same day is the best.
#This is how we categorize the data on ranking base.
```

```python
#I want to get in order as per the ranking.

df.sort_values('shipmode')
```

```python
df.sort_values('shipmode', ascending=True)
```

```python
df.sort_values('shipmode', ascending=False)
```

```python
df[df.shipmode > '2nd Class']
```
