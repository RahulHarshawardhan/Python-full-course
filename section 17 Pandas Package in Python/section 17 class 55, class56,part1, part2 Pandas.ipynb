```python
# When the column name is different then how to do the joins. The joining method is different.

stu_details = {"stu_id": [1,2,3,4,5],
              'stu_name':['A','B','C','D','E'],
              'Region':['N','S','W','E','S']}

stu_marks = {'st_id': [1,2,3,4,6],
            'sub': ["Py","Tab","R","Py","Py"],
           'Marks': [35,32,34,37,38] }

```

```python
stu_details = pd.DataFrame(stu_details)
```

```python
stu_details
```

```python
stu_marks = pd.DataFrame(stu_marks)
```

```python
stu_marks
```

```python
pd.merge(stu_details, stu_marks, left_on='stu_id', right_on = 'st_id' , how = "inner") 
# When column name is diferent we have left_on, right_on command.
```
