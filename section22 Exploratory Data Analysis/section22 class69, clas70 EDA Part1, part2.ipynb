# Exploratory Data Analysis(EDA)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv('iris.csv')
```

```python
iris
```

```python
#Once we get the data what we have to do , that is EDA. and we will do it here.
# I want to predict which type species on sepal length, sepal width, petal length, petal width.
# entire row of (sepal_length	sepal_width	petal_length	petal_width) is vector or an array in python terms.
#species is a dependent variable/ output/ class/ response/ Label
#(sepal_length	sepal_width	petal_length	petal_width) are Features/ Variable/ independent variable/ input.
```

```python
#How many records and variables are there? 
print(iris.shape)
```

```python
#What all columns we have in the dataset?
print(iris.columns)
```

```python
iris["species"].value_counts() # These types of data is called balanced data.
```

```python
#What is balanced data?
# All types of species have exactly 50 datas. So these type of data is called balanced data.

#Non Balance data:
# when there is difference in datas between the categories is imbalance data.
```

```python
print("Means:")
print(np.mean(iris_setosa['petal_length']))

#Mean with an outlier
#print(np.mean(np.append(iris_setosa['petal_length'], 50)));# here we add outlier 50 to the setosa petal_length,
print(np.mean(iris_virginica['petal_length']))              #mean drastically changed.
print(np.mean(iris_versicolor['petal_length'])) 

print('\nStd-dev:')
print(np.std(iris_setosa['petal_length']))
print(np.std(iris_virginica['petal_length']))
print(np.std(iris_versicolor['petal_length']))
```

```python
print('\nMedians:')
print(np.median(iris_setosa['petal_length']))
print(np.median(iris_virginica['petal_length']))
print(np.median(iris_versicolor['petal_length']))

print("\nQuantiles:")
print(np.percentile(iris_setosa['petal_length'], np.arange(0,100,25)))
print(np.percentile(iris_virginica['petal_length'], np.arange(0,100,25)))
print(np.percentile(iris_versicolor['petal_length'], np.arange(0,100,25)))


print("\n90th percentiles:")
print(np.percentile(iris_setosa['petal_length'],90))
print(np.percentile(iris_virginica['petal_length'], 90))
print(np.percentile(iris_versicolor['petal_length'], 90))

from statsmodels import robust
print("\nMedian Absolute Deviation")
print(robust.mad(iris_setosa["petal_length"]))
print(robust.mad(iris_virginica["petal_length"]))
print(robust.mad(iris_versicolor["petal_length"]))
```
