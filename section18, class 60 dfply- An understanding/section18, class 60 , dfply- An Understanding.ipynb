```python
tips >> mutate(total_final_bill = X.total_bill + X.tip) >> select(columns_from('total_bill')) >> head(3)
```

```python
#I have to do per person calculation from total final bill.

tips.rename(columns = {'size': 'count_per'},inplace = True)
```

```python
tips
```

```python
tips >> mutate(total_final_bill= X.total_bill + X.tip, per_head_cost = ((X.total_bill+X.tip)/X.count_per))>> select(columns_from('total_bill')) >> head(3)
```

```python
# transmute - will give only the columns created.
tips >> transmute(total_final_bill = X.total_bill+X.tip, per_head_cost = ((X.total_bill + X.tip)/X.count_per)) >> head(3) 
```

```python
# Summarized data

tips>> summarise(total_bill_mean= X.total_bill.mean(), total_bill_std = X.total_bill.std())
```

```python
tips >> group_by('sex') >> summarise(total_bill_mean= X.total_bill.mean(), total_bill_std = X.total_bill.std())
```

```python
import numpy as np

tips >> summarise_each([np.mean, np.var], X.total_bill, X.tip)
```

```python
tips >> summarise_each([np.mean, np.var], X.total_bill, 'tip') #Here giving the column name
```

```python
tips >> summarise_each([np.mean, np.var], X.total_bill, 1)# here giving the index.
```
