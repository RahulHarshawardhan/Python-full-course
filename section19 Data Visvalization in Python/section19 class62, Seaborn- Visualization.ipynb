```python
import pandas as pd
import seaborn as sns
```

```python
car = pd.read_csv('Car_sales.csv')
```

```python
car
```

```python
# I want to know how my fuel efficiency is disributed. 
```

```python
#IN sns there is aplot called distribution plot.

sns.distplot(car['Fuel efficiency'])
```

```python
car.isnull().sum()
```

```python
sns.distplot(car['Sales in thousands'], kde = True)
# Here the majority of the value is at the left side. 
# Majority of the values is having low sales values. 
# This type of the chart is Right skewed charts, Outliers are at right side.
#Desity in graph is called probability value.
```

```python
sns.distplot(car['Sales in thousands'], kde = False) # kde is kernel density estimate
#kde = False gives counts in number.
```

```python
#scatter plots.
#sns we have 2 kinds of plot- 1) Regression(regplot) plot and 2)LM plot Linear Model plot.

sns.regplot(data = car, y ='Fuel efficiency', x = 'Fuel capacity')
# The line is called regression line.
```

```python
sns.regplot(data = car, y ='Fuel efficiency', x = 'Fuel capacity', fit_reg= False)
# fitreg = False means regression line will  be removed.
```

```python
sns.lmplot(data = car, y ='Fuel efficiency', x = 'Fuel capacity')
```

```python
sns.lmplot(data = car, y ='Fuel efficiency', x = 'Fuel capacity', fit_reg= False)
# Lmplot shown the same plot as regplot, but the difference is that we can differentiate the colour of dots as per the
# vehicle type.
```

```python
sns.lmplot(data = car, y ='Fuel efficiency', x = 'Fuel capacity', fit_reg= False, hue= 'Vehicle type')
#hue is not present in the regplot.
```

```python
sns.lmplot(data = car, y ='Fuel efficiency', x = 'Fuel capacity', fit_reg= False, col= 'Vehicle type')
#col gives separate graph for both the vehicle type.
#colis not present in regplot
```

```python
sns.jointplot(data = car, x = 'Fuel capacity', y = 'Fuel efficiency')
```

```python
sns.jointplot(data = car, x = 'Fuel capacity', y = 'Fuel efficiency', kind = 'kde')
```
