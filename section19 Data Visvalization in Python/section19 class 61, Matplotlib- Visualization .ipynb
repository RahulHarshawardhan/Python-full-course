```python
import matplotlib.pyplot as plt 
```

```python
year = [2015,2016,2017,2018,2019] # This year here is not considered as date time format. This is integer.
sales = [4356,4432,3333,2322,4323]
profit = [453,223,344,433,233]
```

```python
# I am looking to create a chart what is sales over the time. 

plt.plot(year,sales, 'red')# Any thoing on the left hand side of the comma is x-axis and right side is y- axis.

plt.title('Year Vs Sales Analysis')
plt.xlabel('Years')
plt.ylabel('Sales')
plt.show()  # To remove the caption.
```

```python
# We can show the profit also in the same table.
plt.plot(year,sales, 'red')# Any thing on the left hand side of the comma is x-axis and right side is y- axis.
plt.plot(year, profit, 'blue')
plt.title('Year Vs Profit Sales Analysis')
plt.xlabel('Years')
plt.ylabel('Sales')
plt.show()  # To remove the caption.
```

```python
# To create 2 charts one for sales and one for profit over the years.


plt.subplot(2,1,1) # Here 2 means i want 2 rows (sales and profit chart), 1 means 1 column, 
#and 1 means i am creating 1 st chart
plt.plot(year,sales, 'red')
plt.title("Sales by Year")
plt.xlabel("Years")
plt.ylabel('Sales')


plt.subplot(2,1,2) # Here last 2 means i am creating second chart
plt.plot(year,sales, 'blue')
plt.title("Profit by Year")
plt.xlabel("Years")
plt.ylabel('Profit')
# Chart will come one below another.
plt.tight_layout() # Tocorrect the overlap and clutternes
```

```python
# To create 2 charts one for sales and one for profit over the years.


plt.subplot(1,2,1)# Here 1 means i want 1 rows (sales and profit chart), 2 means 2 column, 
#and 1 means i am creating 1 st chart
plt.plot(year,sales, 'red')
plt.title("Sales by Year")
plt.xlabel("Years")
plt.ylabel('Sales')


plt.subplot(1,2,2)# here last 2 means creating 2nd chart
plt.plot(year,sales, 'blue')
plt.title("Profit by Year")
plt.xlabel("Years")
plt.ylabel('Profit')
# Chart will come one below another.
plt.tight_layout() # To correct the overlap and clutterness
```

```python
# To create 2 charts one for sales and one for profit over the years.


plt.subplot(2,2,1)
plt.plot(year,sales, 'red')
plt.title("Sales by Year")
plt.xlabel("Years")
plt.ylabel('Sales')


plt.subplot(2,2,2)
plt.plot(year,sales, 'blue')
plt.title("Profit by Year")
plt.xlabel("Years")
plt.ylabel('Profit')
# Chart will come one below another.
plt.tight_layout() # To correct the overlap and clutterness
```

```python
import pandas as pd
```

```python
car =pd.read_csv('Car_sales.csv')
```

```python
# I want to plot a chart which will have manufacturer name amd mean 4 year resale value.

for group, data in car.groupby('Manufacturer'): # group is iterating on manufacturer, data =4 year sales
    print(group,data['4-year resale value'].mean())
```

```python
car['Manufacturer']
```

```python
plt.figure(figsize= (10,7))
for group, data in car.groupby('Manufacturer'): # group is iterating on manufacturer, data =4 year sales
    plt.barh(group,data['4-year resale value'].mean())
```

```python
plt.figure(figsize= (30,7))
for group, data in car.groupby('Manufacturer'): # group is iterating on manufacturer, data =4 year sales
    plt.bar(group,data['4-year resale value'].mean())
```

```python
# I want to check how many vehicle type is there.

car['Vehicle type'].unique()
```

```python
car
```

```python
# I want to how many manufacturer are passenger and car type. 

vt_dict = dict()
for i in set(car['Vehicle type']):
    vt_dict[i] = car[car['Vehicle type']==i].Manufacturer.count()
vt_dict 

plt.bar(vt_dict.keys(), vt_dict.values())
```

```python
plt.figure(figsize=(10,8))
for group, data in car.groupby('Manufacturer'):
    plt.barh(group,data['4-year resale value'].mean())
    plt.barh(group,data['Price in thousands'].mean(), left = data['4-year resale value'].mean())
```

```python
#pie chart

vt_dict = dict()
for i in set(car['Vehicle type']):
    vt_dict[i] = car[car['Vehicle type']==i].Manufacturer.count()
vt_dict 

plt.pie(vt_dict.values(), labels = vt_dict.keys())
plt.show()
```
