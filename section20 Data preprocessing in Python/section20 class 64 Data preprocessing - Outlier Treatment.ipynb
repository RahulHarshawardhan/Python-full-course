```python
def outlier_treatment(dataframe):
    Q1 = dataframe.quantile(0.25)
    Q3 = dataframe.quantile(0.75)
    IQR = Q3-Q1
    df_rem_out = dataframe[~((dataframe < (Q1-1.5*IQR)) | (dataframe > (Q3 + 1.5*IQR))).any(axis = 1)]
    return df_rem_out
```

```python
outlier_treatment(df)
```
